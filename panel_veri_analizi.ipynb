```python
excel_file = 'akaryakit-fiyatlari.xlsx'

# İlk sayfayı yükleme
fiyat_data = pd.read_excel(excel_file, sheet_name='Akaryakıt_Fiyatları')
enflasyon_data = pd.read_excel(excel_file, sheet_name='enflasyon')
```

```python
enflasyon_data.head()
```

```python
enflasyon_data['Enflansyon (TÜFE) oranı'] = enflasyon_data['Enflansyon (TÜFE) oranı'].str.replace(' ', '')  # Boşlukları kaldır
enflasyon_data['Enflansyon (TÜFE) oranı'] = enflasyon_data['Enflansyon (TÜFE) oranı'].str.replace(',', '.').astype(float)  # Virgülleri noktaya dönüştür ve float'a çevir
```

```python
fiyat_data['Fiyat'] = fiyat_data['Fiyat'].str.replace(' TL/LT', '')  # Boşlukları kaldır
fiyat_data['Fiyat'] = fiyat_data['Fiyat'].str.replace(',', '.').astype(float)
```

```python
fiyat_data.head()
```

##excel verilerinin panel dataya çevirilmesi

```python

panel_data = fiyat_data.pivot_table(index='Şehir', columns='Tarih')
panel_data.head()
```

```python
panel_data.info()
```

```python
fiyat_data.columns
```

##Pedroni Panel Eşbütünleşme Analizi

```python
# PanelOLS modelini kullanarak Pedroni Panel Eşbütünleşme Analizi uygulama
model = PanelOLS(panel_data['Fiyat'], panel_data.drop(columns=['Fiyat']))
result = model.fit()
```
